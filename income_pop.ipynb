```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

## Cleaning of data
#### Data retrieved from https://tablebuilder.singstat.gov.sg/table/CT/17779

```python
data = pd.read_csv("17779.csv")
```

```python
data #header is using unnamed
```

```python
 data= data[9:40] #take rows corresponding to the areas of residences
data.columns = data.iloc[0] #set the 1st row taken to be the header instead of unnamed
data = data[1:] 
data.head()
 
```

```python
list(data.columns)
```

```python
new_df = data.dropna(axis=1, how='all') #remove the last column of Na values
```

```python
new_df
```

```python
df2 = new_df.copy(deep=True)
df2.iloc[:,1:] =df2.iloc[:,1:].astype(int) # change dataframe from string to int
```

```python
df2.dtypes
```

```python
df2.iloc[:,2:] = df2.iloc[:,2:].apply(lambda x: x/x.sum() * 100, axis=1) #calculate the % for each category 
```

```python
df2
```

```python
df2.drop('Total', inplace=True, axis=1) #drop the column labelled as 'Total'
```

```python
df2.drop(labels=10, inplace=True, axis=0) #drop the row labelled at "Total"
```

```python
df2
```

## Key findings

```python
df2.plot.bar(x='Planning Area of Residence',y='No Employed Person')
```

```python
df3 = df2.set_index('Planning Area of Residence')
west_areas = pd.concat([df3.iloc[3:9,],df3.iloc[12:14,]])
west_areas
for index, row in west_areas.iterrows(): # iterate through each row
    plt.plot(row, label=index)
plt.legend(loc='center left',bbox_to_anchor=(1.0, 0.5)) #shifts legend to the side
plt.xticks(rotation=90)
plt.title("Income in Residential Areas located in the West")
plt.xlabel("Income")
plt.ylabel("Percentage of population in residential area")
plt.show()
```

### Bukit Merah has the highest concentration of people in lower income groups while Bukit Timah has the lowest concentration of people in lower income groups. Therefore, the data implies that more focus can be placed on Bukit Merah to provide more assistance to people in the lower income groups.
